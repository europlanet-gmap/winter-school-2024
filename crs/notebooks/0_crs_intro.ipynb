## Symbols

Excerpt from rom Snyder (1987): 

<div class="alert alert-block alert-info">  

$a =$ semimajor axis of ellipsoid

$b =$ semiminor axis of ellipsoid

$e =$ eccentricity of the ellipsoid $=(1-b^2/a^2)^{1/2}$

$x =$ rectangular coordinate: distance to the right of the vertical line (Y axis)  passing through the origin or center of a projection (if negative, it is distance to the left). In practice, a "false" x or "false easting'' is frequently added to all values of x to eliminate negative numbers

$y =$ rectangular coordinate: distance above the horizontal line (X axis) passing through the origin or center of a projection (if negative, it is distance below). In practice, a "false" y or "false northing" is frequently added to all values of y to eliminate negative numbers.

$h =$ relative scale factor along a meridian of longitude. (For general perspective projections, h is height above surface of ellipsoid.)

$k =$ relative scale factor along a parallel of latitude


$\lambda = $ longitude east

$\lambda _{0} =$ longitude east of the central meridian of the map, or of the origin of the rectangular coordinates (for west longitude, use a minus sign). 

$\phi = $ north geodetic or geographic latitude (if latitude is south, apply a minus sign).

$\phi _{g} =$ north geocentric / planetocentric latitude (if latitude is south, apply a minus sign).

$\phi _{1} =$ single standard parallel on cylindrical or conic projections; latitude of central point on azimuthal projections.
    
$\theta _{1} =$ angle of intersection between meridian and parallel.
    
$R =$ radius of the sphere, either actual or that corresponding to scale of the map.
    
$\omega =$ maximum angular deformation at a given point on a projection.
    
</div>

## Latitude types

Planetocentric vs. Planetographic/Planetodetic (Geocentric vs. Geodetic)

<img src="https://proj.org/en/9.3/_images/geocentric_latitude.png">

Direct formula,  eq. 3-28 in Snyder (1987), in radians (see also [in 2023 winter school](https://github.com/europlanet-gmap/winter-school-2023/blob/main/crs/notebooks/map-projections.ipynb)):

**Please note: below $e$ is not Euler's number, but e intended as eccentricity of the ellipsoid $=(1-b^2/a^2)^{1/2}$, see Snyder (1987)**.


$$e = \sqrt{1 - \frac{b^2}{a^2}}$$

The geocentric / planetocentric latitude is expressed as: 

$$\phi_{g} = arctan[(1 -e^2)tan(\phi)]$$

Inverse formula:

$$\phi = arctan(tan[\phi_{h})/(1-e^2)]$$

see also:

* https://proj.org/en/9.3/operations/conversions/geoc.html
* https://en.wikipedia.org/wiki/Geodetic_coordinates
* e.g. https://www.oc.nps.edu/oc2902w/coord/coordcvt.pdf
* https://www.iogp.org/wp-content/uploads/2019/09/373-07-02.pdf

When $a=b=c$, i.e. in the case of a sphere, planetocentric and planetographci latitudes coincide. In the case one uses a spheroid as reference (e.g. a biaxial IAU), they do differ.

e.g. on a certain point on - say - Vesta ($a=289000, b=229000$ meters)

### Calulating -graphic/-centric latitudes

Using both proj4 and by hand, following Snyder (1987)

````python
# Import pyproj
from pyproj import CRS, Transformer
import math

# input as sphere
crs_input = CRS.from_proj4('+proj=lonlat +a=289000 +b=229000')

# output as biaxial ellipsoid (using the two end members of - say - of Vesta)
crs_output = CRS.from_proj4('+proj=lonlat +a=289000 +b=229000 + geoc')

transformer = Transformer.from_crs(crs_input, crs_output)

# Define some input coordinates as (longitude, latitude) tuples in degrees

print("With Proj --> at the equator (planetocentric lat=0): ", transformer.transform(-50, 0)[1])
print("With Proj --> at the pole (planetocentric lat=90): ", transformer.transform(-50, 90)[1])
print("With Proj --> halfway (planetographic lat=45) planetocentric is: ", transformer.transform(-50, 45)[1], "\n")

a=289000.0
b=229000.0

e = math.sqrt(1 - (b**2/a**2))
# second eccentricity - see also https://www.iogp.org/wp-content/uploads/2019/09/373-07-02.pdf
# e_prime = e * math.sqrt(a**2/b**2)

planetographic = 45.0
planetographic_radians = math.radians(planetographic)
print("Maually --> planetographic (source) in degrees is: ", planetographic)
print("Maually --> planetographic (source) in radians is: ", planetographic_radians)

# in radians, using eccentricity
planetocentric =  math.atan( (1-e**2) * math.tan(planetographic_radians) ) # use planetographic_radians instead of planetographic
# in degrees
planetocentric_degrees = math.degrees(planetocentric) 
print("Maually --> planetocentric (destination) in radians is: ",planetocentric)
print("Maually --> planetocentric (destination) in degrees is: ",planetocentric_degrees)
````

## Equidistant cylindrical projection

Those are used for the simplicity of their formulas, for later below use. Others are documented in Snyder (1987)

### Direct formulas (x,y metric from lat, lon degrees)

Snyder's (1987) equations 12-1 to 12-4:

$ x = R (\lambda - \lambda _{0}) cos \phi _{1}$

$y = R \phi$

$h = 1$

$k = cos \phi _{1} / cos \phi$

### Inverse formulas (lat, lon degrees, from x, y metric)

Snyder's (1987) equations 12-5, 12-6

$\phi = y / R$

$\lambda = \lambda _{0} + \frac{x}{R cos \phi _{1}}$

--- 

## Sinusoidal Projection

### Direct formulas (x,y metric from lat, lon degrees)

Snyder's (1987) equations 30.1 to 30.5

$ x = R (\lambda - \lambda _{0}) cos \phi$

$y = R \phi$

$h = [1 + (\lambda - \lambda _{0})^2 sin^2 \phi]^{1/2}$

$k = 1.0$

$\theta' = arcsin(1/h)$

$\omega = 2 arctan 1/2(\lambda - \lambda _{0})sin\phi$

### Inverse formulas (lat, lon degrees, from x, y metric)

Snyder's (1987) equations 30-6, 30-7

$\phi = y / R$

$\lambda = \lambda _{0} + \frac{x}{R cos \phi}$

````python
import geopandas as gpd
import pandas as pd
import numpy as np
from shapely.geometry import Point

# Define range and spacing of longitude and latitude values
lon_min = -170
lon_max = 150
lat_min = -80
lat_max = 85
lon_spacing = 30
lat_spacing = 30

gdf = []

# Create arrays of evenly spaced longitude and latitude values
lon_values = np.linspace(lon_min, lon_max, num=int((lon_max-lon_min)/lon_spacing)+1)
lat_values = np.linspace(lat_min, lat_max, num=int((lat_max-lat_min)/lat_spacing)+1)

# Create two-dimensional grids from longitude and latitude arrays
lon_grid, lat_grid = np.meshgrid(lon_values, lat_values)

# Combine longitude and latitude coordinates into tuples
coords = list(zip(lon_grid.flatten(), lat_grid.flatten()))

# Convert tuples into Point objects
points = [Point(coord) for coord in coords]

# Create pandas dataframe from list of Point objects
df = pd.DataFrame({'geometry': points})

# Convert pandas Series into GeoSeries and set geometry column of geodataframe
gdf = gpd.GeoDataFrame(df)
gdf['geometry'] = gpd.GeoSeries(gdf['geometry'])
gdf = gdf.set_geometry('geometry')

# Set coordinate reference system (CRS) of geodataframe (optional)
gdf.crs = 'IAU_2015:30100' # callisto 

# Print or save geodataframe
gdf.plot()
````

````python
import requests
import shutil
import os

url = "https://astropedia.astrogeology.usgs.gov/download/Moon/LRO/LROC_WAC/thumbs/Moon_LRO_LROC-WAC_Mosaic_global_1024.jpg"
output_file = "./moon.jpg"
with requests.get(url, stream=True) as response, open(output_file, 'wb') as out_file:
    shutil.copyfileobj(response.raw, out_file)

plt.figure(figsize=(12, 6))
ax = plt.axes(projection=projection)
ax.set_global()
ax.gridlines(draw_labels=True)

# this would work for polygons but NOT for points ;-|
# ax.add_geometries(gdf.geometry, data_crs) # for Lat/Lon data.

ax.scatter([gdf.geometry.x for point in points],
           [gdf.geometry.y for point in points],
           transform=ccrs.Geodetic(),
           facecolor='yellow')

img = plt.imread(output_file )
img = img[::-1]
ax.imshow(img, origin='lower', transform=ccrs.PlateCarree(), extent=[-180, 180, -90, 90])


plt.title('Moon Orthographic', fontsize=20)
plt.show()
os.remove(output_file)

````
